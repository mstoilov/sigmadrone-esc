```python
import numpy as np
import sigmadrive_ext as sd
from matplotlib.pyplot import plot

pid = sd.PidController(0.49, 0.01, 0.025, 0.1)

t = np.arange(0, 10, 0.1)
K = np.array([[i] for i in np.arange(1,11,2)])
R = np.sum(1 + 1/K*np.sin(K*t), axis=0)
P = []

for r in np.nditer(R):
    pid.Input(pid.Output() - r, 0.1)
    P.append(-pid.Output())

plot(t, R)
plot(t, P)
```
