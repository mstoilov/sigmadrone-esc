```python
import numpy as np
import sigmadrive_ext as sd
from matplotlib.pyplot import plot

a = 0.1
b = 0.9
c = 0.0

dT = 0.1
pid = sd.PidController(0.49, 0.005, 0.025, 0.0)
#pid = sd.PidController(1.9, 0.0, 0.15, 0.0)

t = np.arange(0, 5, dT)
K = np.array([[i] for i in np.arange(1,1001,2)])
R = np.sum(1/K*np.sin(K*t), axis=0) + 0
P = []
I = []
D = []

for r in np.nditer(R):
    c = c*a + b*pid.Output();
    pid.Input(-(r - c), dT)
    P.append(pid.P)
    I.append(pid.I)
    D.append(pid.D)
    

plot(t, R)
plot(t, -(np.array(P) + np.array(D)))
#plot(t, I)
#plot(t, D)
```
